# adding batting summary table

```python
with open (r"C:\Users\91948\OneDrive\Desktop\projects\cricket analytics\t20_json_files\t20_wc_batting_summary.json") as f:
    data = json.load(f)
    
    all_records = []
    
    for rec in data:
        all_records.extend(rec['battingSummary'])
        
df_batting = pd.DataFrame(all_records)
df_batting.head()  #creating single list where all the matches are recorded
```

```python
df_batting['match'].value_counts().sum()
```

 creating  out and not out new column using lamba function

```python
df_batting['out/not_out']=df_batting.dismissal.apply(lambda x : 'out' if len(x) > 0 else 'not_out' )
df_batting.head()
```

```python
df_batting['out/not_out'].value_counts()
```

```python
#droping dismissal column 
df_batting.drop(columns = ['dismissal'],inplace = True)
```

```python
df_batting.columns
```

```python
#removing special characters 
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x : x.replace ('â€',''))
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x : x.replace ('\xa0',''))
```

```python
df_batting.head(30)
```

 now we can'not see that special character it has been removed

 we are going to merge and  batting column tabel into one

```python
match_ids_dict = {}

for index, row in df_match.iterrows():
    key1 = row['team1'] + 'vs' + row['team2']
    key2 = row['team2'] + 'vs' + row['team1']
    
    match_ids_dict[key1] = row['match_id']
    match_ids_dict[key2] = row['match_id']

match_ids_dict
```
